```python
#Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
"""
Type 1 error occurs when we reject a null hypothesis when it is actually true.
Ex:Suppose a pharmaceutical company is testing a new drug for its effectiveness in curing a disease. The null hypothesis in this case would be that 
   the drug is not effective. If the company rejects the null hypothesis based on a small sample size or insufficient statistical evidence, it may 
   lead to the drug being approved and released to the market even though it may not actually be effective.
   
Type 2 error occurs when we fail to reject a null hypothesis when it is actually false.
Ex:Continuing from the previous example, suppose the pharmaceutical company conducted a large clinical trial to test the drug's effectiveness. 
   If the company fails to reject the null hypothesis despite the drug being effective, it may result in the drug not being approved and released 
   to the market, which would be a Type 2 error. This scenario is also called a false negative because a positive effect was missed due to the 
   failure to reject the null hypothesis.
"""
```

```python
#Q4: Explain Bayes's theorem with an example.
"""
It states that the probability of a hypothesis (H) given some observed evidence (E) is proportional to the probability of the evidence given the 
hypothesis, multiplied by the prior probability of the hypothesis. 

In mathematical terms: P(H | E) = P(E | H) * P(H) / P(E)
where P(H | E) is the probability of the hypothesis given the evidence,
      P(E | H) is the probability of the evidence given the hypothesis, 

Ex: Suppose you are a doctor trying to diagnose a patient with a rare disease. The disease affects only 1% of the population, and there is a test
    that can detect the disease with 99% accuracy (i.e., if a person has the disease, the test correctly identifies it 99% of the time, and if a 
    person does not have the disease, the test correctly identifies it 99% of the time).
    If you administer the test to a patient and it comes back positive, what is the probability that the patient actually has the disease?


Let H be the hypothesis that the patient has the disease, and let E be the evidence that the test is positive.

P(H) = 0.01 (the prior probability that the patient has the disease)
P(E | H) = 0.99 (the probability that the test is positive given that the patient has the disease)
P(~H) = 0.99 (the prior probability that the patient does not have the disease)
P(E | ~H) = 0.01 (the probability that the test is positive given that the patient does not have the disease)
Using the formula for Bayes's theorem, we can calculate the probability that the patient actually has the disease given that the test is positive:

P(H | E) = P(E | H) * P(H) / [P(E | H) * P(H) + P(E | ~H) * P(~H)]

P(H | E) = 0.99 * 0.01 / [0.99 * 0.01 + 0.01 * 0.99] = 0.50
"""
```

```python
#Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
"""
A confidence interval is a statistical measure that provides a range of values within which a population parameter, such as a population mean or 
proportion, is likely to lie.

Here is an example of how to calculate a confidence interval for a population mean using a t-distribution:
Suppose we want to estimate the average height of all students in a university. We randomly sample 50 students and measure their heights. 
The sample mean height is 170 cm and the standard deviation of the sample heights is 5 cm. We want to calculate a 95% confidence interval for the
population mean height.

1. Determine the level of confidence: We want a 95% confidence interval, which means that there is a 95% chance that the true population mean falls
   within the interval.
2. Calculate the standard error: The standard error is a measure of the variability of the sample mean. It is calculated as the standard deviation of
   the sample divided by the square root of the sample size. In this case, the standard error is 5 / sqrt(50) = 0.71 cm.
3. Calculate the margin of error: The margin of error is the distance from the sample mean to the upper or lower bound of the confidence interval. 
   It is calculated by multiplying the standard error by the appropriate critical value from the t-distribution. The critical value is determined by 
   the level of confidence and the degrees of freedom, which is equal to the sample size minus one. For a 95% confidence interval with 49 degrees of 
   freedom, the critical value is 2.009. Therefore, the margin of error is 2.009 x 0.71 = 1.43 cm.
4. Calculate the confidence interval: The confidence interval is calculated by adding and subtracting the margin of error to the sample mean. In this
   case, the 95% confidence interval is 170 +/- 1.43, or 168.57 to 171.43 cm.

Therefore, we can say with 95% confidence that the true population mean height of all students in the university falls between 168.57 and 171.43 cm.
"""
```

```python
#Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence.
#    Provide a sample problem and solution.
"""
According to research of publicly-traded companies, 60% of the companies that increased their share price by more than 5% in the last three years 
replaced their CEOs during the period.At the same time, only 35% of the companies that did not increase their share price by more than 5% in the same
period replaced their CEOs. Knowing that the probability that the stock prices grow by more than 5% is 4%, find the probability that the shares of a 
company that fires its CEO will increase by more than 5%.

Before finding the probabilities, you must first define the notation of the probabilities.

P(B)=(the probability that the CEO is replaced)
P(A) =0.04 (the probability that the stock price increases by 5%)
P(B|A)=0.60 (the probability of the CEO replacement given the stock price has increased by 5%)
P(~A)=0.96 (the probability that stock prise doesnot increase by 5%)
p(B|~A)=0.35 (probability that CEO replaced given that stock price doesnt increase)

P(A|B)= the probability of the stock price increases by 5% given that the CEO has been replaced
P(A|B)= P(A)*P(B|A)/{P(A)*P(B|A)+P(~A)*p(B|~A)}
      = 0.04*0.6/{0.04*0.6 + 0.96*0.35}
      = 0.067 or 6.7%
      
Thus, the probability that the shares of a company that replaces its CEO will grow by more than 5% is 6.67%.
"""
```

```python
#Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where 
#    a larger sample size would result in a smaller margin of error.
"""
The margin of error in a confidence interval is the range of values that represents the level of uncertainty in the estimate based on a sample. 
It is the maximum distance between the sample estimate and the true population parameter, within a certain level of confidence.

The margin of error is calculated based on the sample size, the level of confidence, and the variability of the data. The larger the sample size, 
the smaller the margin of error, assuming that all other factors are constant. This is because a larger sample size reduces the sampling error, 
which is the error that arises due to chance variations in the sample.

For example, suppose a researcher wants to estimate the proportion of people in a city who support a new policy initiative. The researcher takes a 
random sample of 1000 people and finds that 60% of them support the policy. Using a 95% confidence level, the margin of error for this estimate is 
±3.1%. This means that the true proportion of people who support the policy in the population is expected to be between 56.9% and 63.1%.

Now suppose the researcher takes a larger sample of 2000 people instead of 1000. Using the same 95% confidence level, the margin of error 
for this new estimate would be ±2.2%. This means that the true proportion of people who support the policy in the population is expected to be 
between 57.8% and 62.2%.

Therefore, increasing the sample size in this scenario would result in a smaller margin of error and a more precise estimate.
"""
```

```python
#Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation 
#     of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.
"""
n=50
x=6
sd=2.5
u=0

Null hypothesis: The true population mean weight loss with the drug is equal to 0 (no weight loss).
Alternative hypothesis: The true population mean weight loss with the drug is less than 0 (significant weight loss).

t= (x-u)*sqrt(n)/sd
 = (6-0)/2.5*sqrt(50)
 = 16.97
 
The degrees of freedom  = n - 1 = 49.
p-value for one-tailed test with significance level of 0.05 is 1.6766

as 16.97>1.6766 we reject the null hypothesis.The new weight loss drug had really worked.
"""
```

```python
#Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of
#     people who are satisfied with their job.
"""
sample size is n = 500,
sample proportion is p = 0.65. 

We want to calculate the 95% confidence interval, so α = 0.05 and Zα/2 = 1.96.

To calculate the confidence interval for the proportion of people who are satisfied with their job, we can use the following formula:

CI = p ± Zα/2 * sqrt(p*(1-p)/n)

where:
p is the sample proportion
Zα/2 is the critical value from the standard normal distribution for a given confidence level (α/2), where α = 1 - confidence level
n is the sample size

CI = 0.65 ± 1.96 * sqrt(0.65*(1-0.65)/500)
CI = 0.65 ± 0.041
CI = (0.609, 0.691)

Therefore, we can say with 95% confidence that the true proportion of people who are satisfied with their job is between 60.9% and 69.1%.
"""
```
