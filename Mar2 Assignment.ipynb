```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Matplotlib Assignment question 2")
plt.show()
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
"""
The subplot() function is a method in the matplotlib library in Python that is used to create subplots within a single plot. It is particularly
useful when you want to display multiple plots together in a single figure.
"""
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2,2,figsize=(10,5))

axs[0,0].plot(x, y1)
axs[0,0].set_title('Line 1')

axs[0, 1].plot(x, y2, 'r')
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x, y3, 'g')
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('LIne 4')

fig.suptitle('Four Line Plots')

plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of a set of data that shows the distribution of the data based on 
five summary statistics: the minimum value, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum value.

Box plots are commonly used in data analysis and statistical inference to summarize the distribution of a dataset and to identify potential outliers.
They are particularly useful for comparing multiple datasets side-by-side and for identifying differences in the distribution of the data, such as
differences in the median or the spread of the data.
"""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data=[box1,box2]

plt.boxplot(data)
plt.show()
```
