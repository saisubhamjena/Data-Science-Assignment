```python
#Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

from scipy import stats

x=50
sd=5
n=100

std_error=sd/n**0.5

C_I=stats.t.interval(0.95,n-1,loc=x,scale=std_error)


print(f"The 95% confidence interval is ({C_I[0]:.2f}, {C_I[1]:.2f})")
```

```python
#Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% 
#    blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

import numpy as np
from scipy import stats

obs = np.array([12,14,20,8,10,15])
exp = np.array([.2, .2, .2, .1, .1, .2])* obs.sum()

chi,p=stats.chisquare(obs,exp)

print(f"Chisquare statistics: ({chi:.2f})")
print(f"P-Value: ({p:.2f})")

alpha = 0.05
if p < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
#Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square 
#    statistic of 15.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2


df = 10

x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)

chi_sq = chi2.pdf(x, df)

fig, ax = plt.subplots()
ax.plot(x, chi_sq, 'r-', lw=2, label='chi-square pdf')

mask = x >= chi2.ppf(0.975, df)
ax.fill_between(x[mask], chi_sq[mask], 0, alpha=0.5, color='gray', label='p-value')

ax.set_xlabel('Chi-square statistic')
ax.set_ylabel('Probability density')
ax.set_title(r'$\chi^2$ Distribution with 10 Degrees of Freedom')
ax.legend(loc='best')
plt.show()
```

```python
#Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval
#    for the true proportion of people in the population who prefer Coke.

n=1000
prefered_coke=520
p=prefered_coke/n
z=2.576

std_error=(p*(1-p)/n)**0.5
margin_error=z*std_error

lower_bound = p - margin_error
upper_bound = p + margin_error

print(f"The 99% confidence interval for the proportion of individuals who smoke is: {lower_bound*100:.2f}% to {upper_bound*100:.2f}%")
```

```python
#Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness 
#    of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.


import numpy as np
from scipy import stats

obs=np.array([45,55])
exp=np.array([50,50])

chi,p=stats.chisquare(obs,exp)

print(f"Chisquare statistics: ({chi:.2f})")
print(f"P-Value: ({p:.2f})")

alpha = 0.05
if p < alpha:
    print("Reject null hypothesis ")
else:
    print("Fail to reject null hypothesis ")
```

```python
#Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no)
#    Results are shown in the contingency table below.Conduct a chi-square test for independence to determine if there is a significant association 
#    between smoking status and lung cancer diagnosis.Use a significance level of 0.05.

#Lung Cancer:    Yes     No
#Smoker           60    140
#Non-smoker       30    170 

from scipy.stats import chi2_contingency

observed = [[60, 140], [30, 170]]

chi2_stat, p_val, dof, expected = chi2_contingency(observed)


print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis and conclude that there is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject null hypothesis ")
```
